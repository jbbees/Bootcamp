# Reading a CSV file as a Pandas DataFrame

### Step 1: Create a path to the file

```python
csvpath = Path("../Resources/sales.csv")             # imports file path and stores in variable csvpath
                                                     # if you stored the csv file in the same directory just pass file name  csvpath = Path("sales.csv")
                                                     # 
                                                     # you want to keep code organized into directories
```

# Reading a CSV with no Header

```python
# Without a header in the CSV, Pandas will use the first row as the header

csvpath_noheader = Path("../Resources/sales_no_header.csv")
sales_data = pd.read_csv(csvpath_noheader, header=None)           # when we read-in our file to create a dataframe, pass the header=None to indicate no header
sales_data.head()                                                 # this will dispaly numbers as column names. 
```

```python
# Read the file without a header
```

```python
# Rewrite the column names
columns = ['Name', 'Email', 'Address', 'Zip', 'Credit Card', 'Sale Price']         # set the column names in a list.
sales_data.columns                                                                 # all this displays is the current columns. doesn't update column names.
```

```python
sales_data.columns = columns          # this will inject our list of column names into our dataframe. 
sales_data.head()
```
